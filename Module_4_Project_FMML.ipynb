```python
#TASK 5 # Function for gradient descent
def gradient_descent(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break

    return x_list, slope_list

# Run gradient descent with different learning rates
n = 100
low_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                             multiplier=0.001, precision=0.0001, max_iter=n)

high_gamma = gradient_descent(derivative_func=dg, initial_guess=3,
                              multiplier=0.002, precision=0.0001, max_iter=n)

# Experiment
insane_gamma = gradient_descent(derivative_func=dg, initial_guess=1.9,
                                multiplier=0.25, precision=0.0001, max_iter=n)

#TASK 6 # Plotting
plt.figure(figsize=[20, 5])

# 1 Chart: Derivative dg(x)
plt.subplot(1, 3, 1)
plt.title('Derivative dg(x)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.plot(x_2, dg(x_2), color='orange', linewidth=5, alpha=0.6)

# 2 Chart: Learning Rate Comparison
plt.subplot(1, 3, 2)
plt.title('Learning Rate Comparison', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()

# Plot low_gamma
plt.scatter(low_gamma[0], low_gamma[1], color='red', s=50, alpha=0.5, label='Low Gamma (0.0005)')

# Plot mid_gamma
plt.scatter(mid_gamma[0], mid_gamma[1], color='blue', s=50, alpha=0.5, label='Mid Gamma (0.001)')

# Plot high_gamma
plt.scatter(high_gamma[0], high_gamma[1], color='green', s=50, alpha=0.5, label='High Gamma (0.002)')

# Plot insane_gamma
plt.scatter(insane_gamma[0], insane_gamma[1], color='purple', s=50, alpha=0.5, label='Insane Gamma (0.25)')

plt.legend()

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)
plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(1.5, 2)
plt.ylim(-0.5, 0.5)

# Plot low_gamma
plt.scatter(low_gamma[0], low_gamma[1], color='red', s=50, alpha=0.5, label='Low Gamma (0.0005)')

# Plot mid_gamma
plt.scatter(mid_gamma[0], mid_gamma[1], color='blue', s=50, alpha=0.5, label='Mid Gamma (0.001)')

# Plot high_gamma
plt.scatter(high_gamma[0], high_gamma[1], color='green', s=50, alpha=0.5, label='High Gamma (0.002)')

# Plot insane_gamma
plt.scatter(insane_gamma[0], insane_gamma[1], color='purple', s=50, alpha=0.5, label='Insane Gamma (0.25)')

plt.legend()

plt.show()
```

```python
# Function for gradient descent with cost reduction tracking
def gradient_descent_with_cost_reduction(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    cost_reduction_list = [f(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        cost_reduction_list.append(f(new_x) - f(previous_x))

        if step_size < precision:
            break

    return x_list, slope_list, cost_reduction_list


n = 100
low_gamma_with_cost_reduction = gradient_descent_with_cost_reduction(derivative_func=dg, initial_guess=3,
                                                                      multiplier=0.0005, precision=0.0001, max_iter=n)


plt.figure(figsize=[15, 5])

# 1 Chart: Cost Reduction
plt.subplot(1, 2, 1)
plt.title('Cost Reduction per Iteration', fontsize=17)
plt.xlabel('Iteration', fontsize=16)
plt.ylabel('Cost Reduction', fontsize=16)
plt.grid()
plt.plot(range(len(low_gamma_with_cost_reduction[2])), low_gamma_with_cost_reduction[2], color='blue', linewidth=3)

# 2 Chart: Learning Rate Comparison
plt.subplot(1, 2, 2)
plt.title('Learning Rate Comparison', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()


plt.scatter(low_gamma[0], low_gamma[1], color='red', s=50, alpha=0.5, label='Low Gamma (0.0005)')

plt.legend()

plt.show()
```
